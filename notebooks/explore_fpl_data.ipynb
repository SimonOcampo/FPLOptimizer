```python
import json
import pandas as pd

# Load the player/team data
with open("../data/bootstrap-static.json") as f:
    data = json.load(f)
```

```python
# 'elements' contains the list of all players
players = pd.DataFrame(data["elements"])
players.head()

# Simplify the view
players_simple = players [[
        "first_name", "second_name", "team", "now_cost", "total_points", "minutes", "form", "selected_by_percent"
]].copy()
players_simple.head()
```

```python
# Load team info
teams = pd.DataFrame(data["teams"])

# Create a hashmap from team ID to team name
team_id_map = dict(zip(teams["id"], teams["name"]))

# Add team name to player data
players_simple["team_name"] = players["team"].map(team_id_map)

# Show updated player info
players_simple[["first_name", "second_name", "team_name", "now_cost", "total_points"]].head()
```

```python
# Load position info
positions = pd.DataFrame(data["element_types"])

# Map element type 1-5 to position names
position_map = dict(zip(positions["id"], positions["singular_name"]))

# Add position to player data
players_simple["position"] = players["element_type"].map(position_map)

# Preview the result
players_simple[["first_name", "second_name", "team_name", "position", "now_cost", "total_points"]].head()
```
